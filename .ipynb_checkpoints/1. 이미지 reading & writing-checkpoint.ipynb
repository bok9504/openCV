```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# openCV의 함수를 활용한 이미지파일 읽고 화면에 표시하기

def showImage():
    imgfile = 'images/model.png'
    img = cv2.imread(imgfile, cv2.IMREAD_COLOR) # 이미지 읽기(컬러로)
    
    cv2.namedWindow('model', cv2.WINDOW_NORMAL) # 크기조정 가능한 윈도우창 생성
    cv2.imshow('model', img) # 윈도우창에 보이기
    cv2.waitKey(0) # 입력 대기시간
    cv2.destroyALLWindows() # 생성한 윈도우 제거
    
showImage()
```

```python
def showImage():
    imgfile = 'images/model.png'
    img = cv2.imread(imgfile, cv2.IMREAD_UNCHANGED) # 원본
    cv2.imshow('model', img)
    
    k = cv2.waitKey(0) & 0xFF # 사용자 키보드 입력값을 k에 대입
    
    if k == 27: # ESC 입력시 프로그램 종료(아스키 코드)
        cv2.destroyAllWindows()
    elif k == ord('c'): # c 입력시 다른 파일로 저장
        cv2.imwrite('images/model_copy.jpg', img) # img write
        cv2.destroyAllWindows()
        
showImage()
```

```python
def showImage():
    imgfile = 'images/model.png'
    img = cv2.imread(imgfile, cv2.IMREAD_GRAYSCALE) # 흑백
    
    plt.imshow(img, cmap='gray', interpolation='bicubic')
    plt.xticks([])
    plt.yticks([])
    plt.title('model')
    plt.show()
    
showImage()
```
